```python
# 2
# Create a function to plot rankings over time for 1 roller coaster
def time_line(df, rollercoaster_name):
    # isolating rollercoaster name
    rollercoaster_data = df[df['Name'] == rollercoaster_name]  # Filter the DataFrame based on roller coaster name
    # creating lineplot
    sns.lineplot(x='Year of Rank', y='Rank', data=rollercoaster_data, hue='Park', marker='o', palette='deep')
    plt.title('Rank of rollercoaster ' + rollercoaster_name + ' in multiple parks')
    plt.xlabel('Year of Rank')
    plt.ylabel('Rank')
    plt.legend(title='Park', loc='center left', bbox_to_anchor=(1, 0.5))
    plt.gca().invert_yaxis()  # Invert the y-axis so that top rank 1 is at the top
    plt.show()
    plt.clf()
    
# Create a plot of El Toro ranking over time
time_line(rollercoaster, 'El Toro')
```

```python
# 3
# Create a function to plot rankings over time for 2 roller coasters
def rollercoaster_ranking(df, rollercoaster_park_data, *rollercoaster_names):
    # isolate rollercoaster name and park name
    for name, park in rollercoaster_park_data:
        rollercoaster_data = df[(df['Name'] == name) & (df['Park'] == park)]  # Filter the DataFrame based on roller coaster name and park
        # create line plot
        sns.lineplot(x='Year of Rank', y='Rank', data=rollercoaster_data, marker='o', label=f'{name} ({park})')
    # add details to line plot
    plt.title('Rank of rollercoasters')
    plt.xlabel('Year of Rank')
    plt.ylabel('Rank')
    plt.legend(title='Rollercoaster (Park)', loc='center left', bbox_to_anchor=(1, 0.5))
    plt.gca().invert_yaxis()  # Invert the y-axis so that top rank 1 is at the top
    plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))  # Set the y-axis tick locator to display only integers
    plt.show()
    plt.clf()

#  Create a plot of El Toro and Boulder Dash ranking over time at specific locations
rollercoaster_park_data = [('El Toro', 'Six Flags Great Adventure'), ('Boulder Dash', 'Lake Compounce')]
rollercoaster_ranking(rollercoaster, rollercoaster_park_data, 'El Toro', 'Boulder Dash')
```

4. Write a function that will plot the ranking of the top `n` ranked roller coasters over time as lines. Your function should take a number `n` and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

   For example, if `n == 5`, your function should plot a line for each roller coaster that has a rank of `5` or lower.

```python
# 4
# Create a function to plot top n ranking rollercoasters over time
def top_ranking(df,n):
    # isolate top ranking roller coaster
    top_n = df[(df.Rank <= n)]
    # creating lineplot for top_n rollercoasters
    sns.lineplot(x='Year of Rank', y='Rank', data=top_n, hue='Name', marker='o', palette='deep')
    plt.title('Best ranking rollercoasters')
    plt.xlabel('Year of Rank')
    plt.ylabel('Rank')
    plt.gca().invert_yaxis()  # Invert the y-axis so that top rank 1 is at the top
    plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))  # Set the y-axis tick locator to display only integers
    plt.legend(title='Rank', loc='center left', bbox_to_anchor=(1, 0.5))
    plt.show()
    plt.clf()

# Create a plot of top 3 rankings over time
top_ranking(rollercoaster, 3)
```

8. Write a function that creates a pie chart that compares the number of operating roller coasters (`'status.operating'`) to the number of closed roller coasters (`'status.closed.definitely'`). Your function should take the roller coaster DataFrame as an argument. Make sure to include informative labels that describe your visualization.

   Call your function with the roller coaster DataFrame.

```python
# 8
# Create a function to plot a pie chart of status.operating
# status type
print(roller_data.status.unique()) 
# ['status.operating','status.relocated','status.closed.definitely','status.retracked','status.closed.temporarily','status.unknown', 'status.construction','status.announced','status.rumored']
def pie(df):
    # isolating operating status
    operating = df[df.status == 'status.operating']
    operating_len = len(operating)
    # cisolating closed definitely status
    closed_definitely = df[df.status == 'status.closed.definitely']
    closed_definitely_len = len(closed_definitely)
    # proportions of both
    prop = [operating_len,closed_definitely_len] 
    # plotting pie
    plt.pie(prop, labels=['operating','closed definitely'], startangle=0, autopct='%1.1f%%')
    plt.axis('equal')
    plt.title('Proportion of Rollercoaster Status')
    plt.show()
    plt.clf()

# Create pie chart of roller coasters
pie(roller_data)
```

9. `.scatter()` is another useful function in matplotlib that you might not have seen before. `.scatter()` produces a scatter plot, which is similar to `.plot()` in that it plots points on a figure. `.scatter()`, however, does not connect the points with a line. This allows you to analyze the relationship between two variables. Find [`.scatter()`'s documentation here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).

   Write a function that creates a scatter plot of two numeric columns of the roller coaster DataFrame. Your function should take the roller coaster DataFrame and two-column names as arguments. Make sure to include informative labels that describe your visualization.

```python
# 9
# Create a function to plot scatter of any two columns
def scatter(df, var1, var2, i):
    # create scatterplot using actual values from the DataFrame
    plt.scatter(x=df[var1], y=df[var2], s=3, color=i)  # Extract values from the DataFrame directly
    plt.title('Relationship between ' + var1 + ' and ' + var2)
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.show()

# Plot speed vs length
scatter(roller_data, 'speed', 'length', 'orange')
# Plot speed vs loops
scatter(roller_data, 'speed', 'num_inversions', 'green')

# limit height to 200m
heights2 = roller_data[roller_data['height'] <= 200]

# Create a function to plot scatter of speed vs height
def scatter_height(df, var1, var2, i):
    # create scatterplot using actual values from the DataFrame
    plt.scatter(x=df[var1], y=df[var2], s=3, color=i)  # Extract values from the DataFrame directly
    plt.title('Relationship between ' + var1 + ' and ' + var2)
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.show()
# Create a scatter plot of roller coaster height by speed
scatter_height(heights2, 'height', 'speed', 'blue')
# Create a scatter plot of roller coaster height by length
scatter_height(heights2, 'height', 'length', 'maroon')
```

```python
# Manufacturers specialties
# unique manufacturers
#display(roller_data.manufacturer.unique()) # ['Vekoma', 'Zamperla', 'William J. Cobb', 'Soquet', 'Pinfari', etc...

#function plots bar graph for desired 'builder' and shows averages for
#each of height, speed, length, and number inversions
#function accepts a dataframe of coaster data and a string with manufacturer name
def manufacturer(df, builder):
    builder_data = df[df.manufacturer == builder]
    avg_height = builder_data.height.mean()
    avg_speed = builder_data.speed.mean()
    avg_length = builder_data.length.mean()
    avg_inversions = builder_data.num_inversions.mean()
    #Only wanted to show '<1' for data as it is somewhat meaningless to 
    #have .24 inversion, for example
    if avg_inversions < 1:
        avg_inversion_text = '< 1'
    else:
        avg_inversion_text = str(round(avg_inversions))
    ax = plt.subplot()
    plt.bar(range(4), [avg_height, avg_speed, avg_length, avg_inversions])
    #Plotting calculated averages on each bar
    ax.text(0, avg_height + 5, str(round(avg_height,2)), horizontalalignment='center', fontsize=12)
    ax.text(1, avg_speed + 5, str(round(avg_speed,2)), horizontalalignment='center', fontsize=12)
    ax.text(2, avg_length + 5, str(round(avg_length,2)), horizontalalignment='center', fontsize=12)
    ax.text(3, avg_inversions + 5, avg_inversion_text, horizontalalignment='center', fontsize=12)
    ax.set_xticks(range(4))
    ax.set_xticklabels(['Avg Height(m)', 'Avg Speed (kmh)', 'Avg Length (km)', 'Avg # of Inversions'], rotation=90)
    plt.title('Averages on ' + str(len(builder_data)) + ' Coasters for Manufacturer: ' + builder)
    plt.show()
    plt.clf()

# plot
manufacturer(roller_data, 'Vekoma') 
```

```python
#function will plot a line graph of the parks with the 20 highest, longest,
#or fastest coaster... function accepts 'height','length','speed' 
#and 'num_inversions' for col_name
def top20(df, col_name):
    #find heightest coaster for each park, ignore nans and 0s
    park_data = df[df[col_name] != 0].groupby('park')[col_name].max().dropna().reset_index()
    #Too much data! Plot only top 20 parks with highest coasters
    top_20_park_data = park_data[col_name].nlargest(n=20).reset_index()
    #Now we have a df with height data, also need a matching list of park names
    #for the x labels
    top_20_park_names = []
    for row in range(len(top_20_park_data)):
        top_20_park_names.append(park_data.iloc[int(top_20_park_data['index'][row])].park)
    ax = plt.subplot()
    #print(list(zip(top_20_park_names, top_20_park_data[col_name])))
    plt.plot(range(len(top_20_park_data)), top_20_park_data[col_name])
    ax.set_xticks(range(len(top_20_park_data)))
    ax.set_xticklabels(top_20_park_names, rotation = 90)
    plt.ylabel(col_name)
    #Pick the appropriate title based on col_name
    if col_name == 'speed':
        plt.title('Top 20 Parks with Fastest Coasters')
    elif col_name == 'height':
        plt.title('Top 20 Parks with Heighest Coasters')
    elif col_name == 'length':
        plt.title('Top 20 Parks with Longest Coasters')
    else: #num_inversions
        plt.title('Top 20 Parks with Coasters having most Inversions')
    plt.show()
    plt.clf()
    
# plot data
top20(roller_data, 'speed')
top20(roller_data, 'height')
top20(roller_data, 'length')
```
